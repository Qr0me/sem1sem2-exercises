```python
import matplotlib.pyplot as plt
day = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps_walked = [5000, 8000, 6000, 10000, 7500, 9000, 7000]
steps_walked_another_week = [6000, 7500, 9000, 8000, 9500, 7000, 8500]

plt.grid(True)
plt.plot(day, steps_walked, "-x")
plt.plot(day, steps_walked_another_week, "o-")
plt.title("Steps walked for each day")
plt.show()
```
